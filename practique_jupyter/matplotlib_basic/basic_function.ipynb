```python
import matplotlib.pyplot as plt
```

```python
plt.plot([1, 2, 3, 4, 5, 6, 7, 8, 9])
plt.ylabel("Number y")
plt.xlabel("Number x")
plt.show()
```

```python
plt.plot([1, 2, 3, 4, 5, 6, 7], [1, 3, 6, 21, 34, 45, 46])#versatile function, build in an arbitrary way taking the given elements
```

```python
plt.plot([1, 2, 3, 4, 5, 6, 7], [1, 3, 6, 21, 34, 45, 46], 'ro')
plt.axis((0, 8, 0, 50))#[xmin, xmax, ymin, ymax]
plt.show()
```

```python
import numpy as np
example = np.arange(0., 7., 0.3)
plt.plot(example, example, 'r--', example, example**2, 'bs', example, example**3, 'g^')#transfer of different styles of a matrix generated with numpy
plt.show()
```

Plotting with keyword strings

```python
data = {'a': np.arange(50),
        'c': np.random.randint(0, 50, 50),
        'd': np.random.randn(50)}
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100

plt.scatter('a', 'b', c='c', s='d', data=data)
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()
```

Plot with categorical variables

```python
fig = plt.figure()  # an empty figure with no Axes
fig, ax = plt.subplots()  # a figure with a single Axes
fig, axs = plt.subplots(2, 2)  # a figure with a 2x2 grid of Axes
# a figure with one axes on the left, and two on the right:
fig, axs = plt.subplot_mosaic([['left', 'right_top'],
                               ['left', 'right_bottom']])
```
